```python
from examples.seismic import Model, plot_velocity
from devito import *
import numpy as np
import matplotlib as plt
from scipy.ndimage import zoom

nx = 369
nz = 375
nbl = 40
nshots = 151
space_order = 8
dtype = np.float32
shape = (nx, nz)
spacing = (25., 8.)
origin = (0., 0.)

v = np.empty(shape, dtype=dtype)
path = "model-375.cwp"
with open(path, "rb") as a:
    v = np.fromfile(a, dtype=dtype).reshape([nx, nz])
v = v / 1000

scale_factor = 1.
v_resampled = zoom(v, zoom=scale_factor, order=3)  
new_spacing = tuple(s * scale_factor for s in spacing)
new_shape = v_resampled.shape

model_resampled = Model(vp=v_resampled, origin=origin, shape=new_shape, spacing=new_spacing,
                        space_order=space_order, nbl=nbl, bcs="damp")

plot_velocity(model_resampled)
```

```python
filter_sigma = (6, 6)
model_smooth = Model(vp=zoom(v_resampled, 1.0, order=0), origin=origin, shape=new_shape,
                     spacing=new_spacing, space_order=space_order, nbl=nbl)
gaussian_smooth(model_smooth.vp, sigma=filter_sigma)

plot_velocity(model_smooth)
```

```python
from examples.seismic import TimeAxis

t0 = 0. 
tn = 5000.
dt = 0.0008
time_range = TimeAxis(start = t0, stop = tn, step = dt)

```

```python
from examples.seismic import RickerSource

f0 = 0.015
src = RickerSource(name='src', grid=model_resampled.grid, f0=f0, npoint=1, time_range=time_range)
src.coordinates.data[0, :] = np.array(model_resampled.domain_size) * .5
src.coordinates.data[0, -1] = 1.0
```

```python
rec_coordinates = np.empty((model_resampled.shape[0], 2))
rec_coordinates[:, 0] = np.linspace(0, model_resampled.domain_size[0], num=model_resampled.shape[0])
rec_coordinates[:, 1] = 1.0

plot_velocity(model_resampled, source=src.coordinates.data, receiver=rec_coordinates[::10, :])
```

```python
from examples.seismic import AcquisitionGeometry

geometry = AcquisitionGeometry(model_resampled, rec_coordinates, src.coordinates.data, t0, tn, f0=f0, src_type='Ricker')
geometry.src.show()
```

```python

from examples.seismic import PointSource

def ImagingOperator(model, image):
    v = TimeFunction(name='v', grid=model.grid, time_order=2, space_order=space_order)
    u = TimeFunction(name='u', grid=model.grid, time_order=2, space_order=space_order, save=geometry.nt)
    
    eqn = model.m * v.dt2 - v.laplace + model.damp * v.dt.T
    stencil = Eq(v.backward, solve(eqn, v.backward))
    
    dt = model.critical_dt
    residual = PointSource(name='residual', grid=model.grid,
                           time_range=geometry.time_axis,
                           coordinates=geometry.rec_positions)    
    res_term = residual.inject(field=v.backward, expr=residual * dt**2 / model.m)
    image_update = Eq(image, image - u * v)

    return Operator([stencil] + res_term + [image_update], subs=model.spacing_map)
```

```python
source_locations = np.empty((nshots, 2), dtype=np.float32)
source_locations[:, 0] = np.linspace(0., model_resampled.domain_size[0], num=nshots)
source_locations[:, 1] = 1

plot_velocity(model_resampled, source=source_locations)
```

```python

image = Function(name='image', grid=model_resampled.grid,space_order=space_order)
op_imaging = ImagingOperator(model_resampled, image)

for i in range(nshots):
    print('Imaging source %d out of %d' % (i+1, nshots))
    geometry.src_positions[0, :] = source_locations[i, :]
    true_d, _, _ = solver.forward(vp=model_resampled.vp)
    smooth_d, u0, _ = solver.forward(vp=model_smooth.vp, save=True)
    v = TimeFunction(name='v', grid=model_resampled.grid, time_order=2, space_order=space_order)
    residual = smooth_d.data - true_d.data
    op_imaging(u=u0, v=v, vp=model_smooth.vp, dt=model_smooth.critical_dt, residual=residual)
```
